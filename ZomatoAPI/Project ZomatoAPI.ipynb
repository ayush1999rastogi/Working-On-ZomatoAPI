# 2.a.2 :Restaurant serving more number of cuisines.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("zomato.csv",engine='python',skipinitialspace=True)
df=df[df.Country_Code==1]
df.dropna(inplace=True)
```

```python
cnt=0
li=[]
for i in df.Cuisines:
    x=i.split(",")
    for j in x:
        cnt=cnt+1
    li.append(cnt)
    cnt=0
```

```python
df['CuisineCount']=li
```

```python
df['CuisineCount'].value_counts()
```

```python
arr=[]
i=1
while(i<9):
    arr.append(df[df['CuisineCount']==i]['Aggregate_rating'].mean())
    i=i+1
arr
```

```python
x_axis =np.arange(1,9)
plt.bar(x_axis,arr)
plt.xlabel("Number of cuisine")
plt.ylabel("Average Rating")
plt.show()
```

From this we infer that, the retaurant with higher ratings have more No. of Cuisines.

# Question 3:Visualization-
Part 1 :Plot the bar graph top 15 restaurants have a maximum number of outlets.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("zomato.csv",engine='python',skipinitialspace=True)
df=df[df.Country_Code==1]
df.dropna(inplace=True)
df
```

```python
df['Restaurant_Name'].value_counts()
```

```python
li_of_top15rest=[]
count=[]
cnt = 0
for i in range(15):
    li_of_top15rest.append(df['Restaurant_Name'].value_counts().index[i])
for j in df['Restaurant_Name'].value_counts():
    if(cnt <15):
        count.append(j)
        cnt=cnt+1
print(li_of_top15rest)
print(count)
```

```python
plt.bar(li_of_top15rest,count)
plt.xticks(rotation = 90)
plt.ylabel("Number of Outlet")
plt.xlabel("Restaurant Name")
plt.show()
```

# Question 3:
Part 4: Plot the pie graph of top 10 cuisines present in restaurants in the USA.

```python
df = pd.read_csv("zomato.csv",engine='python')
df2=df[df.Country_Code==216]
df2.dropna(inplace=True)
cuis=[]
for i in df2.Cuisines:
    ele = i.split(', ')
    for j in ele:
        cuis.append(j)
df2
```

```python
datf = pd.DataFrame(cuis)
```

```python
li_of_top10cuis=[]
count2=[]
cnt = 0
for i in range(10):
    li_of_top10cuis.append(datf[0].value_counts().index[i])
for j in datf[0].value_counts():
    if(cnt <10):
        count2.append(j)
        cnt=cnt+1
print(li_of_top10cuis)
print(count2)
```

```python
plt.figure(figsize=(10,10))
plt.pie(count2 , labels = li_of_top10cuis , autopct='%1.2f%%')
plt.title("Most Popular Cuisines in USA")
plt.show()
```
